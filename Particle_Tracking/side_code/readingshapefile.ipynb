```python
#imports
import geopandas as gpd
# initial position shapfile path
#fp = "/home/samin/data/shapefiles/senseflux_intersectpoint_500m_offshore.shp"
fp = "/home/samin/data/shapefiles/intersectionpoints/GLB_Bdry_buff10km__lh_shore_NOAA_70k_Intersect_snaped.shp"
# Read file using gpd.read_file
data = gpd.read_file(fp)
```

```python
# Rename col 0 to lon and col 1 to lat
data.rename(columns={"POINT_X":"lon", "POINT_Y":"lat"}, inplace=True)
```

```python
# Relocate the first column with 360 coordinates
#relocate the first column with 360 coordinates
data["lon"] = data["lon"] % 360

# add two new coloumns
data["depth"] = 0.0
data["group_id"] = 1
# specify the output directory to save the data
out_path = '/home/samin/data/FVCOME_OUTPUT/simulations/horon/input/converted_senseflux_intersectpoint_5km_offshore_updated.dat'
# Specify the order of columns and save them
out_col_order = ["group_id", "lon","lat", "depth"]
data.loc[:, out_col_order].to_csv(out_path, sep = ' ', index = False, header=False)
```
